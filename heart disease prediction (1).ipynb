```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#data collection and processing
#loading csv data to pandas dataframe
heart_data = pd.read_csv('/content/heart_disease_data.csv')
```

```python
#no of rows and columns in dataset
heart_data.shape
```

```python
#getting more info about the data
heart_data.info()
```

```python
#checking for missing values
heart_data.isnull().sum()
```

```python
#statistical measures of the data
heart_data.describe()
```

```python
#checking the distribution of target variable
heart_data['target'].value_counts()
```

```python
#one is disease and 0 is healthy heart
#splitting the features and target
x = heart_data.drop(columns="target", axis=1)
#if dropping a column axis should be 1 and if dropping row, axis is 0
y = heart_data['target']
print(x)
print(y)
```

```python
#splitting the data into training data and test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, stratify=y, random_state= 2)
#stratify is used because train ma sab 0 value ni huna sakxa test ma sab 1 value
#matra ni huna sakxa so to even out the data, stratify use garne
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
#model training
model = LogisticRegression()
```

```python
#model learning model ma fit garne data
model.fit(x_train, y_train)
```

```python
#model evaluation on how well it is performing
#accuracy score
#on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print(training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
print(test_data_accuracy)
```

```python
#building a predictive system
input_data = (63,1,3,145,233,1,0,150,0,2.3,0,0,1)
#change the input data to a numpy array
input_data_as_numpy_array= np.asarray(input_data)
#reshaping the numpy array as we are predicting for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction= model.predict(input_data_reshaped)
print(prediction)
```
